```python
import matplotlib.pyplot as plt
import numpy as np

f, (ax1, ax2) = plt.subplots(1, 2, width_ratios=[1, 1])
f.set_figheight(5)
f.set_figwidth(14)

time = []
data = []

names = ['postProcessing/fieldAverageCelda1/0/volFieldValue_0.dat']

for name in names:
    with open(name, 'r') as f:
        d = f.readlines()
        for i in d:
            if i[0] != '#':
                k = i.rstrip().split(" ")
                data.append(float(k[-1].strip('\t')))
                time.append(float(k[0]))

data = np.array(data)
time = np.array(time)

ax1.plot(time,data,label='zona 1')

time = []
data = []

names = ['postProcessing/fieldAverageCelda1U/0/volFieldValue_0.dat']

for name in names:
    with open(name, 'r') as f:
        d = f.readlines()
        for i in d:
            if i[0] != '#':
                k = i.rstrip().split(" ")
                data.append(float(k[-1].strip('\t')))
                time.append(float(k[0]))

data = np.array(data)
time = np.array(time)

ax2.plot(time,data,label='zona 1')
# ax1.set_ylim([0, 1])

time = []
data = []

names = ['postProcessing/fieldAverageCelda2/0/volFieldValue_0.dat']

for name in names:
    with open(name, 'r') as f:
        d = f.readlines()
        for i in d:
            if i[0] != '#':
                k = i.rstrip().split(" ")
                data.append(float(k[-1].strip('\t')))
                time.append(float(k[0]))

data = np.array(data)
time = np.array(time)

ax1.plot(time,data,label='zona 2')

time = []
data = []


names = ['postProcessing/fieldAverageCelda3/0/volFieldValue_0.dat']

for name in names:
    with open(name, 'r') as f:
        d = f.readlines()
        for i in d:
            if i[0] != '#':
                k = i.rstrip().split(" ")
                data.append(float(k[-1].strip('\t')))
                time.append(float(k[0]))

data = np.array(data)
time = np.array(time)

ax1.plot(time,data,label='zona 3')


##########


ax1.set_title('Promedio variable en zona de interés')
ax1.set_xlabel('tiempo')
ax1.legend()

ax2.set_title('CoV en zona de interés')
ax2.set_xlabel('tiempo')
ax2.legend()

# plt.savefig('concentracion.png')
plt.show()


```
